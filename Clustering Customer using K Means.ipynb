```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
customer_data = pd.read_csv('C:/Users/DEBASHISH THAKUR/Desktop/Project Customer segmentation/Mall_Customers.csv')
```

```python
customer_data
```

```python
#choosing the annual income column and spendinf score column
X = customer_data.iloc[:,[3,4]].values
```

```python
print(X)
```

```python
#choosing the number of clusters
#WCSS -> Within clusters Sum of squares
```

```python
#wcss value for different number of clusters
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
#plot elbow curve
sns.set()
plt.plot(range(1,11), wcss)
plt.title('Elbow point graph')
plt.xlabel('No. of cluster')
plt.ylabel('wcss')
```

```python
#k = 5 optimum number of clusters
```

```python
#Training the k-means clustering model
kmeans = KMeans(n_clusters = 5, init='k-means++', random_state=42)
Y = kmeans.fit_predict(X)
print(Y)
```

```python
#visualising all the clusters and their centroids
plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0], X[Y==0,1], s=50, c='green', label='Cluster1')
plt.scatter(X[Y==1,0], X[Y==1,1], s=50, c='red', label='Cluster2')
plt.scatter(X[Y==2,0], X[Y==2,1], s=50, c='yellow', label='Cluster3')
plt.scatter(X[Y==3,0], X[Y==3,1], s=50, c='violet', label='Cluster4')
plt.scatter(X[Y==4,0], X[Y==4,1], s=50, c='blue', label='Cluster5')

#plot centroids
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='cyan', label='centroids')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
